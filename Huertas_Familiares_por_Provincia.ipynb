```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
argentina_fp="/content/provincia.json"
provincias=gpd.read_file(argentina_fp)
```

```python
provincias.plot()
```

```python
huertas=pd.read_csv("/content/huertas-familiares-por-provincia.csv")
```

```python
huertas=huertas.rename(columns = {'provincia_id': 'in1'})
```

```python
provincias.in1=provincias.in1.astype("int")
```

```python
prov=pd.merge(provincias, huertas, on ='in1', how ='inner')
```

```python
huertas.head()
```

```python
mapa, ax = plt.subplots(figsize=(10, 15))
#ax.set_title("Argentina - Incidencia Pobreza",  fontdict={'fontsize':20, 'color': '#1273ab'})

prov.plot(column="2018", cmap='autumn',  categorical=True, ax=ax, edgecolor='black')

plt.show()
```
